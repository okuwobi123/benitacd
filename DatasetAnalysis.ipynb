```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("Dataset/heart.csv")

# Set visual style
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 5)

# 1. Age Distribution
plt.figure()
sns.histplot(df['age'], bins=20, kde=True, color='skyblue')
plt.title('Age Distribution of Patients')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

# 2. Gender Distribution
plt.figure()
sns.countplot(data=df, x='sex', palette='Set2')
plt.title('Gender Distribution (0 = Female, 1 = Male)')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.xticks([0, 1], ['Female', 'Male'])
plt.show()

# 3. Heart Disease by Gender
plt.figure()
sns.countplot(data=df, x='sex', hue='target', palette='Set1')
plt.title('Heart Disease by Gender')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.xticks([0, 1], ['Female', 'Male'])
plt.legend(title='Heart Disease', labels=['No (0)', 'Yes (1)'])
plt.show()

# 4. Chest Pain Type Distribution
plt.figure()
sns.countplot(data=df, x='cp', palette='coolwarm')
plt.title('Chest Pain Type (03)')
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')
plt.show()

# 5. Correlation Heatmap
plt.figure(figsize=(12, 8))
corr = df.corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap="YlGnBu", square=True)
plt.title("Feature Correlation Heatmap")
plt.show()
```
